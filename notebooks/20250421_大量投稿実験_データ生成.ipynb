```python
# rootの.envにOpenAIのAPIキーを配置し、pubcom_inputs配下にkankyou_pubcom.csvを配置して動かしてください

import pandas as pd
import json
import os
```

```python
input_dir = "../data/pubcom_inputs"
input_df = pd.read_csv(os.path.join(input_dir, "kankyou_pubcom.csv"))


```

```python
input_df
```

```python
def parse_class(summary: str) -> int:
    class_text = summary.split(" ")[0]
    return int(class_text)
    
```

```python
input_df["class"] = input_df["要約"].apply(parse_class)
input_df["title"] = input_df["要約"].apply(lambda x: " ".join(x.split(" ")[1:]))
```

```python
classes = list(input_df["class"].unique())
```

```python
import openai
from dotenv import load_dotenv

load_dotenv("../.env")

def request_to_openai(
    messages: list[dict],
    model: str = "gpt-4.1",
    is_json: bool = False,
) -> dict:
    openai.api_type = "openai"
    response_format = {"type": "json_object"} if is_json else None
    response = openai.chat.completions.create(
        model=model,
        messages=messages,
        temperature=0,
        n=1,
        seed=0,
        response_format=response_format,
        timeout=30,
    )
    return response.choices[0].message.content
```

```python
AUGMENTATION_PROMPT_TEMPLATE = """与えられたテキストに対して、意味的に類似する同等の文章を{augmentation_num}件生成してください。

# 前提
* LLMによるデータ拡張について実験を行っており、生成されたデータは後の実験で活用されます

# 指示
* 与えられたテキストに対して、意味的に類似し、また完全に同一ではない文章を生成してください
    * なるべく、「AIが自動生成した文章」であることがわからないような文章を生成してください
    * {augmentation_num}件の内容は、なるべくそれぞれ異なる書き方で出力してください
        * 特に、なるべく冒頭の文章や表現はそれぞれ変えるように工夫してください
* 出力例に記載したJSON形式で出力してください

# 出力例
{{
    "texts": ["生成された文章1", "生成された文章2", "生成された文章3", ...]
}}

# 生成元のテキスト
{original_text}
"""
```

```python
# LLMでdata augementationを行うクラスを指定
target_class = 13
# 1件あたりの増加数
augmentation_num = 10


target_df = input_df[input_df["class"] == target_class]

augmented_rows = []
for _, row in target_df.iterrows():
    original_text = row["寄せられた御意見の概要"]
    messages = [
        {"role": "system", "content": AUGMENTATION_PROMPT_TEMPLATE.format(original_text=original_text, augmentation_num=augmentation_num)},
    ]
    response = request_to_openai(messages, model="gpt-4.1", is_json=True)
    print("--------------------------------")
    print(original_text)
    print(response)
    results = json.loads(response)["texts"]
    
    for result in results:
        tmp_row = row.to_dict()
        tmp_row["寄せられた御意見の概要"] = result
        tmp_row["is_augmented"] = True
        augmented_rows.append(tmp_row)


augmented_df = pd.DataFrame(augmented_rows)
concat_df = pd.concat([input_df, augmented_df])
```

```python
concat_df["is_augmented"] = concat_df["is_augmented"].fillna(False)

concat_df.to_csv(os.path.join(input_dir, "kankyou_pubcom_augmented.csv"), index=False)



```
